```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = {}
with open("data/learning.csv", "r") as f:
    category = f.readline().strip().split(",")
    for s in category:
        data[s] = []
    for l in f:
        l = l.strip().split(",")
        for i,s in enumerate(l):
            data[category[i]].append(float(l[i]))
```

```python
for s in category:
    plt.figure()
    plt.title(s)
    if s in [" value-loss"," policy-loss"]:

        new_data = np.array(data[s])
        new_data = new_data - min(-1,np.min(new_data)-1)
        plt.plot(new_data)
        plt.yscale('log')
    else:
        plt.plot(data[s])

plt.show()
```
