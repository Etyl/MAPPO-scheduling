```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = {}
with open("data/learning.csv", "r") as f:
    category = f.readline().strip().split(",")
    for s in category:
        data[s] = []
    for l in f:
        l = l.strip().split(",")
        for i,s in enumerate(l):
            data[category[i]].append(float(l[i]))
```

```python
plt.figure(figsize=(20, 10))
plt.plot(data["episode-return"])
plt.xlabel("Episode")
plt.ylabel("Episodic return")
plt.xlim(0, len(data["episode-return"]))
plt.ylim(0, 100)
plt.show()
```

```python
for s in category:
    plt.figure()
    plt.title(s)
    if s == " value-loss":

        new_data = np.array(data[s])
        new_data = new_data - min(-1,np.min(new_data)-1)
        plt.plot(new_data)
        plt.yscale('log')
    else:
        plt.plot(data[s])

plt.show()
```
